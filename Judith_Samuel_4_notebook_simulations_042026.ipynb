```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from scipy.stats import ks_2samp

# 1. Préparation des données
df_m = df_no_outliers.copy()
df_m['last_purchase_timestamp'] = pd.to_datetime(df_m['last_purchase_timestamp'])
df_m['month'] = df_m['last_purchase_timestamp'].dt.to_period('M')

# Features basées sur votre segmentation (RFM + Satisfaction + Délais)
features = ['recence', 'frequence', 'montant', 'review_score', 'processing_delay_days']
months = sorted(df_m['month'].unique())

# Définition de la période de départ (ex: début 2017 pour avoir assez de volume)
start_month = months[3]
remaining_months = [m for m in months if m >= start_month]

# 2. Modèle de RÉFÉRENCE (T0) - Basé sur 5 clusters
data_ref = df_m[df_m['month'] == start_month][features].dropna()
scaler_ref = StandardScaler()
features_ref_scaled = scaler_ref.fit_transform(data_ref)
kmeans_ref = KMeans(n_clusters=5, random_state=42, n_init=10).fit(features_ref_scaled)

ari_scores = []
ks_pvalues = []

# 3. Simulation temporelle
for m in remaining_months:
    data_current = df_m[df_m['month'] == m][features].dropna()
    if len(data_current) < 5: continue

    # Prédiction avec modèle T0
    current_scaled_with_ref = scaler_ref.transform(data_current)
    labels_pred_from_ref = kmeans_ref.predict(current_scaled_with_ref)

    # Nouveau modèle sur T_actuel
    scaler_cur = StandardScaler()
    current_scaled_own = scaler_cur.fit_transform(data_current)
    labels_true_current = KMeans(n_clusters=5, random_state=42, n_init=10).fit_predict(current_scaled_own)

    # Calcul ARI
    ari = adjusted_rand_score(labels_true_current, labels_pred_from_ref)
    ari_scores.append(ari)

    # Test KS sur le 'montant' (indicateur fort de drift)
    _, p_val = ks_2samp(data_ref['montant'], data_current['montant'])
    ks_pvalues.append(p_val)

# 4. Visualisation des résultats
fig, ax1 = plt.subplots(figsize=(14, 6))

# Courbe ARI
ax1.set_xlabel('Mois')
ax1.set_ylabel('ARI (Stabilité)', color='tab:blue')
ax1.plot([str(m) for m in remaining_months[:len(ari_scores)]], ari_scores, marker='o', color='tab:blue', label='ARI Score')
ax1.axhline(y=0.8, color='red', linestyle='--', label='Seuil de Maintenance (0.8)')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.set_ylim(0, 1.1)

# Courbe P-Value (KS Test)
ax2 = ax1.twinx()
ax2.set_ylabel('P-value (KS Test - Drift)', color='tab:orange')
ax2.plot([str(m) for m in remaining_months[:len(ks_pvalues)]], ks_pvalues, marker='x', color='tab:orange', linestyle=':', label='KS P-Value')
ax2.axhline(y=0.05, color='black', linestyle='-', alpha=0.3, label='Seuil de Drift (0.05)')
ax2.set_yscale('log')
ax2.tick_params(axis='y', labelcolor='tab:orange')

plt.title('Evolution de la stabilité des 5 clusters et détection de Drift')
fig.tight_layout()
plt.xticks(rotation=45)
plt.show()

# Conclusion
last_ari = ari_scores[-1]
print(f"Dernier score ARI mesuré : {last_ari:.4f}")
if last_ari < 0.8:
    print("RECOMMANDATION : Le modèle a divergé. Un réentraînement est nécessaire pour maintenir la pertinence des 5 segments.")
else:
    print("Le modèle est encore stable.")
```

### Analyse de la Stabilité Temporelle (Score ARI)

Le graphique de l'évolution du score **Adjusted Rand Index (ARI)** permet de mesurer la pérennité de notre segmentation dans le temps par rapport à un modèle de référence (T0).

#### Observation des résultats :

*   **Chute brutale de l'ARI :** Le score ARI s'effondre rapidement après le premier mois pour atteindre des valeurs proches de **0.006**. Un score ARI de 1 indiquerait une stabilité parfaite, tandis qu'un score proche de 0 indique que la nouvelle segmentation est totalement différente de la précédente.
*   **Seuil de Maintenance :** La courbe se situe largement en dessous du seuil critique de **0.80**. Cela signifie que les caractéristiques des clients évoluent si vite que les clusters définis à l'instant T0 ne sont plus représentatifs dès les mois suivants.
*   **Test de Drift (P-value KS) :** La courbe orange (P-value du test Kolmogorov-Smirnov sur le montant) montre des fluctuations souvent sous le seuil de **0.05**, confirmant statistiquement un changement de distribution des données (Data Drift).

#### Conclusion et Recommandation :

La segmentation est **extrêmement instable**. Pour que l'équipe marketing puisse exploiter ces segments de manière fiable, un **réentraînement du modèle (clustering) est nécessaire tous les mois**, ou au minimum tous les deux mois, pour capturer les nouveaux comportements d'achat et la saisonnalité.

### Importance des variables par Cluster

Pour interpréter les 5 segments, nous analysons les centroïdes (valeurs moyennes normalisées). Cela permet de voir quelles variables ont le plus de poids dans la définition de chaque groupe.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Récupération des centroïdes (valeurs scalées)
centroids = kmeans_ref.cluster_centers_
centroids_df = pd.DataFrame(centroids, columns=features)
centroids_df.index.name = 'Cluster'

# 2. Visualisation par Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(centroids_df, annot=True, cmap='RdYlGn', center=0)
plt.title('Importance des variables par Cluster (Valeurs centrées-réduites)')
plt.show()

# 3. Interprétation textuelle basée sur les centroïdes
print("Profils types des segments :")
for i in range(len(centroids_df)):
    top_feat = centroids_df.iloc[i].idxmax()
    low_feat = centroids_df.iloc[i].idxmin()
    print(f"- Cluster {i} : Marqué principalement par {top_feat} (max) et {low_feat} (min)")
```

### 📊 Interprétation de la Heatmap des Clusters

#### 🔍 Analyse par Segment :

*   **Cluster 0 : Clients Inactifs (Perdus)**

    *   **Interprétation :** Ce sont des clients dont le dernier achat est très ancien. Ils n'interagissent plus avec la plateforme.
